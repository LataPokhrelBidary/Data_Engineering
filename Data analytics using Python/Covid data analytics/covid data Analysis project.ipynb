```python
covid_df.drop(["Sno","Time", "ConfirmedIndianNational","ConfirmedForeignNational"],inplace = True, axis = 1)
covid_df.head()
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format = '%Y-%m-%d')
covid_df.head()
```

```python
covid_df['Active_cases'] = covid_df['Confirmed']- covid_df['Cured']+covid_df['Deaths']
covid_df.tail()
```

```python
state = pd.pivot_table(covid_df,values = ["Confirmed","Deaths","Cured"],index = "State/UnionTerritory",aggfunc = max)
state["Recovery Rate"]= state["Cured"]*100/state["Confirmed"]
state["Mortality Rate"]= state["Deaths"]*100/state["Confirmed"]
state = state.sort_values(by = "Confirmed", ascending = False)
state.style.background_gradient(cmap = "cubehelix")

```

```python
#top 10 active cases using group by
top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by = ['Active_cases'],
                                                                                                               ascending = False).reset_index()
```

```python
fig = plt.figure(figsize = (26,15))
plt.title("Top 10 states most active cases in India",size = 35)
axis = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_cases",
                   x = "State/UnionTerritory",linewidth = 2,edgecolor = 'red')
```

```python
#Top states with highest deaths
top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'],ascending = False).reset_index()
fig = plt.figure(figsize = (18,5))
plt.title("Top 10 States with most deaths", size = 35)
axis = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths",
                   x = "State/UnionTerritory",linewidth = 2, edgecolor = 'blue')
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```
